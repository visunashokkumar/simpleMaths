```python
from matplotlib import pyplot as plt
import numpy as np
import pandas as pd
```

```python
gap= 5

start= -gap
end= gap
res= 0.05

x = np.arange(start, end, res) # adding number to array from the range of -2 to 2 with the steps of 0.01 (start, end, res)
y = np.arange(start, end, res)

x, y = np.meshgrid(x,y)     # makin a mesh grid

z = (x**2)+(y**2)     #putting the equation z= X^2 + Y^2
```

```python
x,y,z     #checkin the arrays
```

```python
# Simple 3d plot of the equation z= X^2 + Y^2
plt.figure(figsize=(16,16))     #setting the image size
ax=plt.axes(projection='3d')
ax.plot_surface(x,y,z)
plt.show()
```

```python
# Simple 3d plot of the equation z= X^2 + Y^2
fig = plt.figure(figsize=(16,16))     #setting the image size
ax = plt.axes(projection='3d')
surf = ax.plot_surface(x,y,z, rstride=2, cstride=2, cmap=plt.cm.Spectral , linewidth=0.5, antialiased=True) #cmap=gradient(Z)
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
fig.colorbar(surf, shrink=0.5, aspect=5)
plt.show()
```
